```python
import pandas as pd
import numpy as np

# Example: your DataFrame is named clean_data

# 1️⃣ Remove unwanted characters from string columns
clean_data['Name'] = clean_data['Name'].str.replace(r'\W', '', regex=True)
clean_data['Domain'] = clean_data['Domain'].str.replace(r'\W', '', regex=True)
clean_data['Location'] = clean_data['Location'].str.replace(r'\W', '', regex=True)

# 2️⃣ Clean 'Age' column — extract only numbers
clean_data['Age'] = clean_data['Age'].str.extract(r'(\d+)').astype(float)

# 3️⃣ Clean 'Exp' (Experience) column — extract only numbers
clean_data['Exp'] = clean_data['Exp'].str.extract(r'(\d+)').astype(float)

# 4️⃣ Handle missing values (you can use mean, median, or fill with specific values)
clean_data['Age'].fillna(clean_data['Age'].mean(), inplace=True)
clean_data['Location'].fillna('Unknown', inplace=True)
clean_data['Exp'].fillna(clean_data['Exp'].mean(), inplace=True)

# 5️⃣ Optional: fix incorrect city names manually
clean_data['Location'] = clean_data['Location'].replace({'Hyderbad': 'Hyderabad'})

# 6️⃣ Reorder columns (optional)
clean_data = clean_data[['Name', 'Domain', 'Age', 'Location', 'Salary', 'Exp']]

# ✅ Final cleaned DataFrame
print(clean_data)
```
